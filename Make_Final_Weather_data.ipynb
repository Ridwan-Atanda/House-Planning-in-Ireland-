```python
import pandas as pd
df = pd.read_csv("C:\\Users\\k1332\\MapReduce-House-Weather\\Converted-Weather-data.csv")
df.head()
```

```python
print("The Column names of the dataset")
df.columns
```

```python
columns = ['county','maxtp','mintp','rain','wdsp','converted-datetime']
df_sub = df[columns]
df_sub.head()
```

```python
df_sub['converted-datetime'][0][5:7]
```

```python
import numpy as np
lb_list = []
spring = ['03','04','05']
summer = ['06','07','08']
autumn = ['09','10','11']
winter = ['12','01','02']
          
for i in range(len(df_sub)):
    try:
        if str(df_sub['converted-datetime'][i])[5:7] in spring:
              lb_list.append('spring')
        elif str(df_sub['converted-datetime'][i])[5:7] in summer:
              lb_list.append('summer')
        elif str(df_sub['converted-datetime'][i])[5:7] in autumn:
              lb_list.append('autumn')
        elif str(df_sub['converted-datetime'][i])[5:7] in winter:
              lb_list.append('winter')
    except:
        lb_list.append('Nan')
print(len(lb_list))
df_sub['Seasons'] = lb_list
np.unique(list(df_sub['Seasons'].values))
```

```python
df_sub.head()
```

```python
df_sub.dtypes
```

```python
df_sub['maxtp'] = pd.to_numeric(df_sub['maxtp'],errors='coerce')
df_sub['mintp'] df_sub.dtypes= pd.to_numeric(df_sub['mintp'],errors='coerce')
```

```python
df_sub.dtypes
```

```python
df_sub['Average_Temp'] = (df_sub['maxtp'] + df_sub['mintp'])/2
df_sub.head()
```

```python
df_sub.to_csv("final_weather_data.csv")
```
