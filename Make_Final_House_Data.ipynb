```python
df_sub = df_sub.dropna(axis=0, subset=['Decision'])
print(len(df_sub))
```

```python
#convert conditional, grant permission to approved
#convert refuse approval and withdrawn applications into disapproved
target = ['CONDITIONAL','Conditional Permission','C','GRANT PERMISSION','GRANT PERMISSION                        ',
         'GRANT PERMISSION                                  ']
for i in range(len(df_sub)):
    if df_sub['Decision'].iloc[i] in target:
        df_sub['Decision'].iloc[i] = 'approved'
    else:
        df_sub['Decision'].iloc[i] = 'disapproved'
        
np.unique(list(df_sub['PlanningAuthority'].values))
```

```python
np.unique(list(df_sub['Decision'].values))
```

```python
# Make new column as depending the seasons of datetime.
df_sub = df_sub.dropna(axis=0, subset=['ReceivedDate'])
df_sub['ReceivedDate'] = pd.to_datetime(df_sub['ReceivedDate'])
lb_list = []
spring = ['03','04','05']
summer = ['06','07','08']
autumn = ['09','10','11']
winter = ['12','01','02']
          
for i in range(len(df_sub)):
    try:
        if str(df_sub['ReceivedDate'][i])[5:7] in spring:
              lb_list.append('spring')
        elif str(df_sub['ReceivedDate'][i])[5:7] in summer:
              lb_list.append('summer')
        elif str(df_sub['ReceivedDate'][i])[5:7] in autumn:
              lb_list.append('autumn')
        elif str(df_sub['ReceivedDate'][i])[5:7] in winter:
              lb_list.append('winter')
    except:
        lb_list.append('Nan')
print(len(lb_list))
df_sub['Seasons'] = lb_list
np.unique(list(df_sub['Seasons'].values))
```

```python
df_sub = df_sub[df_sub.Seasons != 'Nan']
np.unique(list(df_sub['Seasons'].values))
```

```python
df_sub['DecisionDate'] = pd.to_datetime(df_sub['DecisionDate'])
# Get year and month and date values from the full date time format values
df_sub['DecisionDate'] = df_sub['DecisionDate'].dt.date
df_sub.head()
df_sub.to_csv("df_sub_house.csv")
```

# Final Step For the Data Preparing

```python
# load data for final processing
import pandas as pd
df = pd.read_csv("C:\\Users\\k1332\\MapReduce-House-Weather\\final_house_planning_data.csv")
df.head()
```

```python
# Select necessary columns for the final data.
df_final = df[['PlanningAuthority','Decision','GrantDate','Time of Processing']]
df_final.head()
```

```python
# cut and replace the final numerical values from the stirng values of "Time of Processing"
for i in range(len(df_final)):
    df_final['Time of Processing'][i] = str(df_final['Time of Processing'][i]).split(' ')[0]
df_final.head()
```

```python
# check the types of each columns
df_final.dtypes
```

```python
# convert object type to datatime format as default
df_final['GrantDate'] = pd.to_datetime(df_final['GrantDate'])
```

```python
# convert the data type to the numberic
df_final = df_final.dropna(subset=['Time of Processing'])
try:
    df_final['Time of Processing'].astype(str).astype(int)
except:
    pass
```

```python
df_final.dtypes
```

```python
idx_list = []
for i in range(len(df_final)):
    if float(df_final['Time of Processing'][i]) >= 0:
        idx_list.append(i)
print(len(df_final))
print(len(idx_list))
```

```python
df_final = df_final.ix[idx_list]
print(len(df_final))
df_final = df_final.reset_index()
df_final.head()
```

```python
df_final = df_final[['GrantDate','PlanningAuthority','Decision','Time of Processing']]
df_final.tail()
```

```python
# save the dataframe to the final csv file
df_final.to_csv("Final_House.csv", index = False)
```
